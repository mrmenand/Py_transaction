## 门面模式 
门面（facade）设计模式：
- 为子系统的一组接口提供统一的接口，并定义一个高级接口来帮助客户端通过更加简单的方式使用子系统  
- 门面解决的问题：如何用单个接口对象来表示复杂的子系统  


门面：
- 它是一个接口，它知道某个请求可以交个那个子系统处理  
- 使用组合将客户端的请求委派给相应的之系统对象 



```python
class Hotelier(object):
    def __init__(self):
        print("Arrange the Hotel ")

    def _isAvailable(self):
        print("Is the Hotel free for the event on the given day")
        return True

    def bookHotel(self):
        if self._isAvailable():
            print("Registered the Booking \n \n")


class Florist(object):
    def __init__(self):
        print("Flower decorations")

    def setFlowerRequirements(self):
        print("Carnations ,Rose and Lilies woule be used \n\n")


class Caterer(object):
    def __init__(self):
        print("Food Arrangements for the Event")

    def setCusine(self):
        print("chinese & Continetal Cuisine  \n\n")


class Musician(object):
    def __init__(self):
        print("Musical Arrangements ")

    def setMusicType(self):
        print("Jazz and  Classic \n\n ")


class EventMgr:

    def __init__(self):
        print("Event Manager : Let me talk to rhe folks \n ")

    def arrange(self):
        self.hotelier = Hotelier()
        self.hotelier.bookHotel()

        self.florist = Florist()
        self.florist.setFlowerRequirements()

        self.caterer = Caterer()
        self.caterer.setCusine()

        self.musician = Musician()
        self.musician.setMusicType()


class You:
    def __init__(self):
        print("You::Whoa ! Marriages Arrangements  ")

    def askEventManager(self):
        print("You :: let contact the event Manager ")
        em = EventMgr()
        em.arrange()

    def __del__(self):
        print("You :: Thanks the Eevent Manager ，all ppreparation done ！")


you = You()
you.askEventManager()

```

## 最少知识原则  
减少对象之间的交互。  
与迪米特法则是一致的：
- 每个单元对系统中其他单元知道的越少越好  
- 单元应该只与其朋友交流 
- 单元不应该知道
